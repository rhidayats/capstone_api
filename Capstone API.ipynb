```python
selling["InvoiceYear"] = selling["InvoiceDate"].dt.year
selling["Weekday"] = selling["InvoiceDate"].dt.day_name()
```

```python
def penjualan(year='', weekday='', country='', genre='',col=[], sortby=''):
    column = []
    equal = []
    condition = []
    if (year=='') & (weekday=='') & (country=='') & (genre==''):
        return selling
    if year!='':
        column.append('InvoiceYear')
        equal.append('==')
        condition.append(int(year))
    if weekday!='':
        column.append('Weekday')
        equal.append('==')
        condition.append(weekday)
    if country!='':
        column.append('Country')
        equal.append('==')
        condition.append(country)
    if genre!='':
        column.append('Genre')
        equal.append('==')
        condition.append(genre)
    
    data_filter = ' & '.join(f'{i} {j} {repr(k)}' for i, j, k in zip(column, equal, condition))
    
    if sortby!='':
        selling.sort_values(by=sortby, inplace=True)
        
    if col==[]:
        return selling.query(data_filter)
    else:
        return selling.query(data_filter)[col]
```
